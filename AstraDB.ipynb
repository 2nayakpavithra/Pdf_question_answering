```python
from langchain.indexes.vectorstore import VectorStoreIndexWrapper
from langchain_community.document_loaders import PyPDFLoader
from langchain_astradb import AstraDBVectorStore
from langchain_community.llms import HuggingFaceHub
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
loader = PyPDFLoader('/content/drive/MyDrive/COMPUTER NETWORKS NOTES.pdf')
```

```python
pages = loader.load_and_split()
len(pages)
```

```python
pages[0]
```

```python
llm = HuggingFaceHub(model_kwargs={"temperature":0.3,"max_length":100}, huggingfacehub_api_token=huggingfacehub_api_token,repo_id="HuggingFaceH4/zephyr-7b-beta",)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=200,
)
texts = text_splitter.split_documents(pages)
```

```python
embeddings_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
vstore = AstraDBVectorStore(
    embedding=embeddings_model,
    collection_name="langchain_pdf_query",
    api_endpoint=ASTRA_DB_API_ENDPOINT,
    token=ASTRA_DB_APPLICATION_TOKEN
)
```

```python
vstore.add_documents(texts)
astra_vector_index = VectorStoreIndexWrapper(vectorstore=vstore)
```

```python
import re
first_question = True
while True:

    query_text = input("\nEnter your question (or type 'quit' to exit): ").strip()

    if query_text.lower() == "quit":
        break
    if query_text == "":
        continue

    print("\nQUESTION: \"%s\"" % query_text)

    raw_answer = astra_vector_index.query(query_text, llm=llm).strip()

    match = re.search(r"Helpful Answer: (.+)", raw_answer, re.DOTALL)
    answer = match.group(1).strip() if match else raw_answer

    print("ANSWER: \"%s\"\n" % answer)

    print("RELEVANT DOCUMENTS BY SCORE:")
    results = vstore.similarity_search_with_score(query_text, k=4)

    seen_documents = set()
    for doc, score in sorted(results, key=lambda x: x[1], reverse=True):
        if doc.page_content not in seen_documents:
            seen_documents.add(doc.page_content)
            print("    [%0.4f] \"%s ...\"" % (score, doc.page_content[:84]))
    print("*******************************************************************")
    first_question = False

```
